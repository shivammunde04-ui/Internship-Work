```python
# write a program to find factorial of a number using while loop
```

```python
num = 5
fact = 1

while num > 0:
    fact = fact * num
    num -=1
print("factorial =",fact)
```

```python
# practice
#increasing
num =1
fact =1

while num<=5:
    fact = fact * num
    num +=1
print("factorial =",fact)
```
