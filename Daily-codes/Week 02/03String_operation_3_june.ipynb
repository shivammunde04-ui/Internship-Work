```python
"""
take a string as user input => string,charater1,character2
use a loop and if-else to check if a specific character is present in the string or not.
get output for both if the character is present and if it is not present.
""";
```

```python
text = input("Enter string:")
ch1 = input("Enter first character:")
ch2 = input("Enter second character:")

found_ch1=False
found_ch2=False

for char in text:
    if char ==ch1:
        found_ch1= True
    if char ==ch2:
        found_ch2= True
            
if found_ch1:
    print(ch1,"is present")
else:
    print(ch1,"is not present")
if found_ch2:
    print(ch2,"is present")
else:
    print(ch2,"is not present")
        
```
