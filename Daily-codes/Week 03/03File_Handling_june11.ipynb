```python
# write into a myfile.txt with context manager
# use write function and 3 separate lines
```

```python
with open("myfile.txt", "w") as f:
    f.write("frodo took the ring.\n")
    f.write("he started his journey to mount doom.\n")
    f.write("sam accompanied him on the journey.\n")
```
