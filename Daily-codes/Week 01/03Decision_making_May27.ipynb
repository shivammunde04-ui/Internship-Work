```python
"""
menu driven ->
take a string from the user
menu -> upper,lowercase ,3 replace a particular part of the string
for third option it should which part to replacr with what
""";
```

```python
print("welcome")
text=input("Enter a string:")

print("Menu =>\n1.upper\n2.lower\n3.replace")
ch= int(input("Enter your choice:"))

if ch==1:
    print(f"{text.upper()}")
elif ch ==2:
    print(f"{text.lower()}")
elif ch ==3:
    r1=input("which part u want to replace:")
    r2=input("what u want to replace it with")
    new= text.replace(r1,r2)
    print(f"result ->{new}")
```
