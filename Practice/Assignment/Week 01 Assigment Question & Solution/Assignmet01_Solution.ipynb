```python
"""2. Accept a temperature in Celsius from the user, convert it to Fahrenheit and Kelvin, and
print all three values with labels.""";
```

```python
celsius = float(input("Enter temperature in celsius"))

fahrenheit = (celsius * 9/5)+32 # we can directly use the formula in a print
kelvin = celsius + 273.15

print("Celsius =",celsius)
print("Fahrenheit =",fahrenheit)
print("Kelvin =",kelvin)
```

```python
"""6. Build a simple calculator — take two numbers and an operator (+, -, *, /) as input,
perform the operation, and print the result. Print an error message for invalid operators or
division by zero.""";
```

```python


num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":          # we can use match operator
    print("Result =", num1 + num2)

elif operator == "-":
    print("Result =", num1 - num2)

elif operator == "*":
    print("Result =", num1 * num2)

elif operator == "/":
    if num2 != 0:
        print("Result =", num1 / num2)
    else:
        print("Error: Division by zero is not allowed.")

else:
    print("Error: Invalid operator.")
```

```python
"""8. Take a positive integer n as input and print all prime numbers from 2 to n""";
```

```python
n = int(input("Enter value of n: "))

for num in range(2, n + 1):
    prime = True

    for i in range(2, num):
        if num % i == 0:
            prime = False
            break

    if prime:
        print(num, end=" ")
```

```python
"""9. Keep accepting numbers from the user until they enter 0. Then print the total count of
numbers entered, their sum, and their average. Do not include 0 in the calculations.""";
```

```python
count = 0
total = 0

while True:
    num = int(input("Enter number (0 to stop): "))

    if num == 0:
        break

    count += 1
    total += num

if count > 0:
    avg = total / count
else:
    avg = 0

print("Count =", count)
print("Sum =", total)
print("Average =", avg)
```

```python
"""10. Take n as input and print a triangle pattern of n rows made of asterisks, where row i
has (2i - 1) stars centered appropriately.""";
```

```python
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```
