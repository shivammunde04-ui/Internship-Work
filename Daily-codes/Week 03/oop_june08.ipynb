```python
class Person:

    def __init__(self,name,age,ht):
        self.name = name
        self.age = age 
        self.height = ht

    def show_info(self):
        return f"name -> {self.name}, age -> {self.age}, height -> {self.height}"
```

```python
p1 = Person('jatin',20,5.8)
p2 = Person('sachin',22,5.9)
```

```python
p1.show_info()
```

# Encapsulation

```python
class Person:

    def __init__(self,name,age,ht):
        self.name = name
        self.__age = age # private variable
        self.height = ht

    def show_info(self):
        return f"name -> {self.name}, age -> {self.__age}, height -> {self.height}"
```

```python
p1 = Person('jatin',20,5.8)
p2 = Person('sachin',22,5.9)
```

```python
p1.show_info()
```

```python
p1.age
```

```python
p1.name
```

```python
"""
Create a class called Employee =>
attributes -> company, name, monthly_salary, job_designation
here all the object of this company will have the same common company
make salary as private variable
create 3 object -> e1,e2,e3

method ->1.method to calculate the anual salary
        2.method to show entire data
""";
```

```python
class Employee:

    company = "TCS"      # Class Variable (same for all employees)

    def __init__(self, name, salary, job_designation):
        self.name = name
        self.__salary = salary      # Private Variable
        self.job_designation = job_designation

    def annual_salary(self):
        return self.__salary * 12

    def show_data(self):
        print("Company :", Employee.company)
        print("Name :", self.name)
        print("Monthly Salary :", self.__salary)
        print("Job Designation :", self.job_designation)
        print("Annual Salary :", self.annual_salary())
        print()



e1 = Employee("Shivam", 30000, "Python Developer")
e2 = Employee("Jatin", 35000, "Tester")
e3 = Employee("Sachin", 40000, "Project Manager")

```

```python
e1.show_data()
e2.show_data()
e3.show_data()
```
