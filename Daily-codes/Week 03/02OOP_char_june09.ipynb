```python
"""
create a class called student
create 2-3 objects having attributes name ,age

create a method which will add a skill(eg.python.ML,java)to the student
where u can view the info of it later (i.e. skills learned by a student)
""";
```

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.skills = [] # Empty List

    def add_skill(self, skill):
        self.skills.append(skill)

    def show_data(self):
        print("Name :", self.name)
        print("Age :", self.age)
        print("Skills :", self.skills)
        
```

```python
s1 = Student("Shivam", 17)
s2 = Student("yash", 18)

# Adding skills
s1.add_skill("Python")
s1.add_skill("Java")

s2.add_skill("ML")

# display Data
s1.show_data()
s2.show_data()
```

# Inheritance

```python
class Person:
    insti = "MITU"

    def __init__(self,age,ht):
        self.age = age
        self.ht = ht

    def show(self):
        print(f"age -> {self.age}, height -> {self.ht}, institute -> {self.insti}")
```

```python
class Ravi(Person):

     def __init__(self,age,ht, sal):
        super().__init__(age,ht)
        self.sal = sal

     def show_sal(self):
          print(f"salary is {self.sal}")
       
```

```python
a = Ravi(25,5.9,57000)
```

```python
a.show_sal()
```

```python
a.show()
```

```python
class Menuapp:

    def __init__(self):
        self.itemlist = []

    def show_menu(self):

        print("Menu\n 1.add item\n2.view item\n3.exit")
        
    def add_items(self):
        item = input("enter the item name :")
        self.itemlist.append(item)

    def view_item(self):
        return self.itemlist

    def run(self):

        while True:

            self.show_menu()

            ch = int(input  (" enter the choice"))

            if ch == 1:
                self.add_items()
            elif ch == 2:
                self.view_item()
            elif ch == 3:
                print("Exit")
                break
            else:
                return ("Invalid choice")
        
```

```python
app = Menuapp()
app.run()
```

```python
class Person:

    def __init__(self,age,name):

        self.__age = age
        self.__name = name

    # getter
    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age


    # setter
    def set_age(self,age):
        self.__age = age


    def __str__(self):
        return f"age -> {self.__age}, name -> {self.__name}"
```

```python
# create a menu driven app for banking simulation
# check balance, deposit money, withdraw money
# (u can add more things as u see fit to make the app better)
```

```python
class BankApp:

    def __init__(self,balance):
        self.balance = balance

    def show_menu(self):

        print("Menu\n1.Check Balance\n2.Deposit Money\n3.Withdraw Money\n4.Exit")

    def check_balance(self):

        print("Current Balance =", self.balance)

    def deposit_money(self):

        amount = float(input("Enter amount : "))
        self.balance += amount

    def withdraw_money(self):

        amount = float(input("Enter amount : "))

        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Insufficient Balance")

    def run(self):

        while True:

            self.show_menu()

            ch = int(input("Enter the choice : "))

            if ch == 1:
                self.check_balance()

            elif ch == 2:
                self.deposit_money()

            elif ch == 3:
                self.withdraw_money()

            elif ch == 4:
                print("Exit")
                break

            else:
                print("Invalid Choice")


```

```python
app = BankApp(10000)
app.run()
```
